```python

import pandas as pd
import matplotlib.pyplot as plt

data_polygons = pd.read_csv("./exported/polygons.csv")

fig = plt.figure(figsize=(10,5))
p = fig.add_subplot()
p.plot(pd.unique(data_polygons.x), data_polygons[data_polygons.group=="operational_profile"].y, color='black', marker='o', label = "Operational Profile")
plt.title('Operational profile test success rate (absolute mass)', fontsize=14)

cmap = plt.get_cmap("tab10")
color = 0
for id, group in data_polygons[data_polygons.group!="operational_profile"].groupby("group"):
    p.plot(group.x, group.y, color=cmap(color), marker='x', label = id)
    p.fill_between(group.x, group.y, step=None, color=cmap(color), alpha=0.2)
    color += 1

p.set_xlabel('workload')
p.set_ylabel('frequency')
p.grid(True)
p.legend()
plt.show()
```
